```python
df = pd.read_csv("Laliga.csv")
```

```python
df.replace('-', 0,inplace=True)
df=df.rename(columns=df.iloc[0])
df=df.drop(df.index[0])
df
```

```python
df.dtypes
df['Debut'] = pd.to_numeric(df['Debut'].str[0:4])
df[(df['Debut']>=1930 ) & (df['Debut']<=1980)]
```

```python
df['Points'] = pd.to_numeric(df['Points'])
df.nlargest(3, 'Points')
```

```python
df['GoalsFor'] = pd.to_numeric(df['GoalsFor'])
df['GoalsAgainst'] = pd.to_numeric(df['GoalsAgainst'])
def goal_diff_count():
    df['goal_diff'] = df['GoalsFor'] - df['GoalsAgainst']
```

```python
goal_diff_count()
df.sort_values('goal_diff', ascending=False).head(1)
```

```python
df.sort_values('goal_diff', ascending=False).tail(1)
```

```python
df['GamesWon'] = pd.to_numeric(df['GamesWon'])
df['GamesPlayed'] = pd.to_numeric(df['GamesPlayed'])
```

```python
df['Winning Percent'] = (df['GamesWon']/df['GamesPlayed'])*100
```

```python
#df.replace(NaN, 0,inplace=True)
df['Winning Percent'].fillna(0,inplace=True)
df['Winning Percent']
```

```python
df.nlargest(5, 'Winning Percent')
```

```python
df['BestPosition'] = pd.to_numeric(df['BestPosition'])
```
